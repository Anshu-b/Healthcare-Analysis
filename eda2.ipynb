```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
glucose_changes_df = pd.read_csv('glucose_changes.csv')
glucose_changes_df
```

```python
top_meals = (
    glucose_changes_df
    .sort_values(by='120_min_change', key=lambda x: x.abs(), ascending=False)
    .drop_duplicates(subset=['Meal_Time'])
    .head(10)
)
top_meals = top_meals[['Meal_Name', '30_min_change', '60_min_change', '90_min_change', '120_min_change']]
print(top_meals)
top_meals.to_csv("top_meals.csv", index=False)
```

```python
top_meals_2 = (
    glucose_changes_df
    .sort_values(by='120_min_change', key=lambda x: x.abs(), ascending=False)
    .drop_duplicates(subset=['Meal_Time'])
    .head(10)
)
top_meals_2 = top_meals_2[['Meal_Name', 'Total_Calories', 'Total_Carbs', 'Total_Sugar', 'Total_Protein', '30_min_change', '60_min_change', '90_min_change', '120_min_change']]
print(top_meals_2)
top_meals_2.to_csv("top_meals2.csv", index=False)
```
